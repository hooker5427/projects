```python
def cut_feature(col_name ,df  ) :
    data =df.copy()
    minv , a,b,c ,maxv  =data[col_name].describe()[ [ 'min' ,'25%' ,'50%' , '75%' ,'max' ] ].values
    up = a - (c-a)*1.5 
    down = c + (c-a) * 1.5
    if up< minv :
        p_cut = [  up, minv ,  a,b,c, down , maxv]
    else:
        p_cut = [minv , up, a,b,c, down , maxv] 
    data[ col_name + "_cut"] =pd.Categorical( pd.cut(data[col_name] , p_cut )).codes
    del data[col_name]
    from sklearn.preprocessing import LabelEncoder
    encoder = LabelEncoder()
    data[ col_name + "_cut"] = encoder.fit_transform( data[ col_name + "_cut"] )
    return data 
```

```python
for  feature in [ 'income' ,'floor_area' ,"height_cm" ,"weight_jin" ,"family_income"] :
    data = cut_feature(  feature , data ) 
```

```python
def plot_count(  feature_names ) :
    plt.figure(  figsize=( 8 , 20 ) ) 
    for i  , feature in enumerate( feature_names  ,start = 1 ) :
        plt.subplot(len( feature_names)  , 1, i  ) 
        plt.legend( "up_left")
        sns.countplot( x= feature ,  hue ="happiness" , data = data   )
```
